```python
#VADER Valence ware Dictionary for Sentiment Reasoning

nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sia=SentimentIntensityAnalyzer()
```

```python
a='This is a good Movie'
sia.polarity_scores(a)
```

```python
b='This was the best, most Awesome movie  ever made!!!!!!!'
sia.polarity_scores(b)
```

```python
c='This was the worst movie I have ever seen'
sia.polarity_scores(c)
```

Adding scores to DataFrame

```python
df['Scores']=df['review'].apply(lambda r:sia.polarity_scores(r))
```

```python
df.head()
```

```python
df['compound']=df['Scores'].apply (lambda c:c['compound'])
```

```python
df.head()
```

Creating pred_label column based on compound_score

```python
df['pred_label']=df['compound'].apply(lambda l:'pos' if l>0 else 'neg')
df.head()
```

Finding Accuracy

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
Acc_sr=accuracy_score(df['label'],df['pred_label'])
Acc_sr
```

```python
print(classification_report(df['label'],df['pred_label']))
```

```python
confusion_matrix(df['label'],df['pred_label'])
```
